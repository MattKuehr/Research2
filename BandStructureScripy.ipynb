```python
import Solver
import numpy as np
```

```python
# -----------------------------------
# (A) Real stack
# -----------------------------------
solver = Solver.BandStructureSolver(
            epsilons     = [3.8, 1.0],
            mus          = [1.0, 1.0],
            thicknesses  = [0.42, 0.58])        # gamma=None by default

q_vals   = np.linspace(0, np.pi/solver.Lambda, 181)
bands_re, bands_im = solver.solve_bands(q_vals, omega_max=6*np.pi)
solver.plot_real_band_structure(q_vals, bands_re)
```

```python
# -----------------------------------
# (B) PT-symmetric stack
# -----------------------------------
gamma = 0.5
solver_PT = Solver.BandStructureSolver(
                epsilons    = [3.8, 1.0],
                mus         = [1.0, 1.0],
                thicknesses = [0.42, 0.58],
                gamma       = gamma)

q_vals   = np.linspace(-np.pi/solver_PT.Lambda, np.pi/solver_PT.Lambda, 241)
bands_re, bands_im = solver_PT.solve_bands(q_vals, omega_max=3*2*np.pi)
solver_PT.plot_complex_band_structure(q_vals, bands_re, bands_im)
```
